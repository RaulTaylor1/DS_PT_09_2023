### Paso 3. Asígnalo a una variable llamada bebidas.

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv'
```

```python
bebidas = pd.read_csv(url, sep=',', index_col='country')
bebidas
```

### Paso 4. ¿Qué continente bebe más cerveza por término medio?

```python
bebidas_grouped = bebidas.groupby('continent')
tot = bebidas_grouped[['beer_servings']].sum()
tot.sort_values(by= 'beer_servings', ascending=False).head(1)
# bebidas[['beer_servings']]
```

```python
med = bebidas_grouped[['beer_servings']].mean()
med.sort_values(by='beer_servings', ascending=False).head(1)
```
