### Paso 3. Asígnalo a una variable llamada usuarios.

```python
usuarios = pd.read_csv(url, sep='|', index_col='user_id')
usuarios
```

### Paso 4. Descubre cuál es la edad media por ocupación

```python
ocupacion = usuarios.groupby('occupation')
ocupacion[['age']].mean()
```
